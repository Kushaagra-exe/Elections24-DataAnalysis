```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python

df = pd.read_csv('Parties.csv')
df
```

```python
url = 'https://results.eci.gov.in/PcResultGenJune2024/index.htm'

response = requests.get(url)
soup = BeautifulSoup(response.content,'html.parser')
```

```python
data = soup.find('table')
data
```

```python
links = {}
links_clean = []
a_tags =data.find_all('a')

# print(a_tags)
for i in a_tags:
    links_clean.append(i.get('href'))
   
for j in range(df.shape[0]):
#         print(df.loc[j]['Parties Names'])
        links[df.iloc[j]['Parties Names']] = links_clean[j]

# links_clean
links
# df.iloc[1]['Parties Names']
```

```python
url = 'https://results.eci.gov.in/PcResultGenJune2024/'+ links['Bharatiya Janata Party - BJP']

resp = requests.get(url)
soup = BeautifulSoup(resp.content,'html.parser')
table = soup.find('table')
```

```python
heads = []

for i in table.find_all('th'):
    heads.append(i.text.strip())
    
heads = heads[1:]
heads_for_combined = ['Party Name']
heads_for_combined.extend(heads)


heads, heads_for_combined
```

```python
candidates_partywise = []
combined_candidates_list = []
for i in links:
    candidates = []
    req = requests.get('https://results.eci.gov.in/PcResultGenJune2024/' + links[i])
    partywise_soup= BeautifulSoup(req.content, 'html.parser')
    tab = partywise_soup.find('table')
    
# appending data into candidates for individual party data 
    
    for j in tab.find_all('tr')[1:]:
        row = j.find_all('td')[1:]
        col = [k.text.strip() for k in row]
        candidates.append(col)
    candidates_partywise.append(candidates)
    
    df = pd.DataFrame(candidates, columns=heads)
    df.to_csv(i+'.csv',index=False)
    candidates2 = []
    
# adding data into combined_candidates_list for combined data
    for j in tab.find_all('tr')[1:]:
        row = j.find_all('td')[1:]
        col = [k.text.strip() for k in row]
        candidates2.append(col)
        candidates2.insert(0,i)
    combined_candidates_list.append(candidates2)
        
```
