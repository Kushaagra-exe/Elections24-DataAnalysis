```python
party_data = soup.find('table').find_all('tr')[1:]
party_data
```

```python
# all the parties
parties_names = []
for i in party_data[:-1]:
    data = i.find('td')
#     print(data)    
    parties_names.append(data.text)
parties_names
```

```python
parties = pd.DataFrame(parties_names, columns = ['Parties Names'])
parties.to_csv('Parties.csv',index=False)
```

```python
# Parties withb their data
party_wise = []
for i in party_data[:-1]:
    data = i.find_all('td')
#     print(data)
    cols =[j.text.strip() for j in data]
    party_wise.append(cols)

#     print(cols)
party_wise
```
