```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
url = 'https://results.eci.gov.in/PcResultGenJune2024/index.htm'

response = requests.get(url)
soup = BeautifulSoup(response.content,'html.parser')
states_dropdown = soup.find('select')
```

```python

state_values2 = []
options = states_dropdown.find_all('option')
for i in options[1:]:

    state_values2.append((i.text.strip() ,'https://results.eci.gov.in/PcResultGenJune2024/partywiseresult-' + i.get('value') + '.htm'))



```

```python
state_values2
```

```python

final_result = []

for name, link in state_values2:
    res = requests.get(link)
    soup = BeautifulSoup(res.content,'html.parser')
    table = soup.find('table')
    anchor_tags = table.find_all('a')
    links = []
    party_wise_links = []
    for tag in anchor_tags:
        link = tag.get('href')
        if link:
            links.append(link)
    party_wise_links = ['https://results.eci.gov.in/PcResultGenJune2024/'+link for link in links]
    for i in party_wise_links:
        win_candidate_list = []
        party_res = requests.get(i)
        party_soup = BeautifulSoup(party_res.content,'html.parser')
        win_candidate_table = party_soup.find('table')
        for row in win_candidate_table.findAll('tr')[1:]:
            cells = row.findAll('td')[1:]
            row_data = [cell.text.strip() for cell in cells]
            row_data.insert(0,name)
            win_candidate_list.append(row_data)
        final_result.append(win_candidate_list)
```

```python
final_result

```

```python
final_cleaned = []

for i in final_result:
    for j in i:
        final_cleaned.append(j)
final_cleaned
```
