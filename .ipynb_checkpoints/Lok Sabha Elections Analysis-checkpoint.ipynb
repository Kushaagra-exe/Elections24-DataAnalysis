```python
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns
```

```python
winner_candidates = pd.read_csv('Scrapped Datasets\winner combined candidates.csv')
state_data = pd.read_csv('Scrapped Datasets\State wise data.csv')
participating_parties = pd.read_csv('Scrapped Datasets\Parties participated.csv')
seats_won_data = pd.read_csv('Scrapped Datasets\party_wise_results.csv')
```

## Combining 2 datasets for better Insights

```python
combined_data = pd.merge(winner_candidates, state_data, on=['Parliament Constituency', 'Winning Candidate', 'Total Votes', 'Margin'])
combined_data
```

## Handling Nan values and chainging datatypes in dataset

```python
combined_data[combined_data['Total Votes'] == '-'] = combined_data[combined_data['Total Votes'] == '-'].replace('-', '0')
combined_data[combined_data['Total Votes'] == '0']
```

```python

combined_data['Total Votes'] = combined_data['Total Votes'].astype(int)
combined_data['Margin'] = combined_data['Margin'].astype(int)
```

```python
combined_data['Total Votes'].dtype , combined_data['Margin'].dtype
```

```python
combined_data.describe()
```

# <u>_Data Insights_</u>


# (1) Total Number of Votes given to Winning Candidates (State Wise)

```python
# group the data by state and sum the total votes
state_votes = combined_data.groupby("State")["Total Votes"].sum()

# create a bar plot
plt.figure(figsize=(10, 6))
state_votes.plot(kind="bar")
plt.title("Total Votes by State")
plt.xlabel("State")
plt.ylabel("Total Votes")


plt.show()
```

# (2) Top 10 Parties with Highest Seats

```python
sns.barplot(x=seats_won_data['Party'][:10], y=seats_won_data['Won'][:10])
plt.title('Top 10 Parties with Highest Seats')
plt.xlabel('Party Name')
plt.ylabel('Total Seats')
plt.xticks(rotation=75)
plt.plot()
```

# (5) Frequency Distribution of Winning Margins

```python

sns.histplot(combined_data['Margin'], bins=20, kde=True)
plt.title('Distribution of Winning Margins')
plt.xlabel('Winning Margin')
plt.ylabel('Frequency')
plt.show()

# State	Parliament Constituency	Winning Candidate	Total Votes	Margin
```

# (6) Scatter Plot for Margins v/s Votes

```python
plt.figure(figsize=(25, 10))
sns.scatterplot(data=combined_data, x=combined_data['Total Votes'], y=combined_data['Margin'], hue=combined_data['Party Name'])
# plt.scatter(combined_data['Total Votes'], combined_data['Margin'], hue = combined_data['Party Name'])
plt.title('Total Votes vs. Winning Margin')
plt.xlabel('Total Votes')
plt.ylabel('Winning Margin')
plt.legend(title='Party')
plt.show()
```

# (7) Highest and lowest Victory Candidate

```python
highest_victory_margin = combined_data.loc[combined_data['Margin'].idxmax()]
lowest_victory_margin = combined_data.loc[combined_data['Margin'].idxmin()]
```

```python
a = (highest_victory_margin, lowest_victory_margin)
```

```python
candidate = []
party = []
margin = []
for i in a:
    candidate.append(i['Winning Candidate'])
    party.append(i['Party Name'])
    margin.append(i['Margin'])
```

```python
candidate,party, margin
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=candidate, y=margin,  hue=party, palette='muted')
plt.title('Candidates with Highest and Lowest Victory Margin')
plt.xlabel('Candidate Name')
plt.ylabel('Victory Margin')
plt.xticks(rotation=45)
plt.show()
```

# (8) Top 10 candidates with highest votes

```python
sorted_data = combined_data.sort_values(by='Total Votes', ascending=False)
sorted_data[:10]
```

```python

sns.barplot(x=sorted_data['Winning Candidate'][:10], y=sorted_data['Total Votes'][:10])
plt.title('Top 10 Leading Candidate by Votes')
plt.xlabel('Candidate')
plt.ylabel('Total Votes')
plt.xticks(rotation=45)
plt.plot()
```

# (9) Top 10 candidates with Lowest votes

```python
sorted_data = combined_data.sort_values(by='Total Votes', ascending=True)
sorted_data[:10]
```

```python
sns.barplot(x=sorted_data['Winning Candidate'][:10], y=sorted_data['Total Votes'][:10])
plt.title('Top 10 Leading Candidate by Votes')
plt.xlabel('Candidate')
plt.ylabel('Total Votes')
plt.xticks(rotation=85)
plt.plot()
```

# (10) Top 10 Parties with Highest Votes

```python
party_wise = sorted_data.groupby('Party Name')['Total Votes'].sum().sort_values(ascending=False)[:10]
party_wise
```

```python

sns.barplot(x=party_wise.index, y=party_wise)
plt.title('Top 10 Parties by Votes')
plt.xlabel('Party Name')
plt.ylabel('Total Votes')
plt.xticks(rotation=75)
plt.plot()
```
